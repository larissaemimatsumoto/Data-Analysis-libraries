```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
```

```python
vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

```python
vendas_df['Imposto'] = vendas_df['Valor Final'] * 0.3
display(vendas_df)
```

```python
vendas_df['Código Venda'] = vendas_df['Código Venda'].fillna(0)
display(vendas_df)
```

```python
vendas_df = vendas_df.dropna()
display(vendas_df)
```

```python
vendas_df = vendas_df.drop('Código Venda', axis=1)
display(vendas_df)
```

```python
 faturamento_produto = vendas_df.groupby('Produto').sum()
display(faturamento_produto)
```

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

```python
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```

```python
transacoes_loja = vendas_df[['ID Loja', 'Gerente']].value_counts()
display(transacoes_loja)
```

```python
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
```
