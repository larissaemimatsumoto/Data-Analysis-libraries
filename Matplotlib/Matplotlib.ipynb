```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 1, 3, 2]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, label = 'dados')
plt.ylabel('Fase')
plt.xlabel('Nível')
plt.title('Pesquisa')
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label = 'dados', linestyle = 'dotted', lw = 2, color = 'm')
plt.ylabel('Fase')
plt.xlabel('Nível')
plt.title('Pesquisa')
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, color = '#0b03fc', label = 'dados')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.show()
```

```python
plt.bar(x, y, color = '#440866', label = 'dados')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, color = '#0b03fc', label = 'dados', marker = '*')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color = 'r', label = 'dados', ls = '--')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.axis(xmin = -1, xmax = 5, ymin = 0, ymax = 4.5)
plt.show()
```

```python
plt.plot(x, y, color = 'g', label = 'dados', ls = '--')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.axis('auto')
plt.show()
```

```python
plt.plot(x, y, color = '#065406', label = 'dados', ls = '--')
plt.xlabel('Nível')
plt.ylabel('Fase')
plt.legend()
plt.axis('square')
plt.show()
```

```python
import pandas as pd
```

```python
medalhas_df = pd.read_excel('Medals.xlsx')
display(medalhas_df)
```

```python
medalhas_win = medalhas_df[['Rank', 'Team/NOC', 'Gold', 'Silver', 'Bronze']]
display(medalhas_win)
```

```python
dfremove = medalhas_win.loc[20:93]
top_medals = medalhas_win.drop(dfremove.index)
display(top_medals)
```

```python
import numpy as np
```

```python
labels = top_medals['Team/NOC']
gold_medals = top_medals['Gold']
silver_medals = top_medals['Silver']
bronze_medals = top_medals['Bronze']

y = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots(figsize=(15, 25))
gold = ax.barh(y - width-0.45*-1, gold_medals, width, label='Gold', color = '#a89003')
silver = ax.barh(y - width*0, silver_medals, width, label='Silver', color = '#948e8a')
bronze = ax.barh(y - width-0.045*1, bronze_medals, width, label='Bronze', color = '#b55202')

ax.set_ylabel('Medals quantity')
plt.rc('ytick', labelsize=14)
plt.rc('xtick', labelsize=14) 
ax.set_title('Tokio 2020')
plt.rc('axes', labelsize=14)
plt.rc('legend', fontsize=14)

ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.legend()

ax.bar_label(gold, padding=3)
ax.bar_label(silver, padding=3)
ax.bar_label(bronze, padding=3)

plt.grid()
plt.tight_layout()

plt.savefig('Tokio 2020.png')

plt.show()
```
